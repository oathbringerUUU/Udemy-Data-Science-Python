# Wykres liniowy

```
import matplotlib
matplotlib.__version__
```

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
fig = plt.figure()
fig.show()
```

```
_ = plt.plot([0, 1], [0, 3])
```

```
_ = plt.plot([0, 1, 2], [0, 3, 2], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend()
```

```
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

_ = plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

# Tworzenie wielu wykresów

```
fig = plt.figure(figsize=(12, 8))

_ = plt.subplot(131)
_ = plt.plot([0, 1, 2], [3, 2, 5])
_ = plt.subplot(132)
_ = plt.plot([0, 1, 2], [3, 1, 4], 'r--')
_ = plt.subplot(133)
_ = plt.plot([0, 1, 2], [3, 1, 4], 'ro')
```

# Wykresy słupkowe

```
plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 1.2, 5.2])
```

```
plt.bar(x=['a', 'b', 'c', 'd'], height=[1.2, 3.2, 1.2, 5.2])
```

```
plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 1.2, 5.2])
```

# Wykresy punktowe

```
x1 = np.random.randn(300)
x2 = np.random.randn(300)

_ = plt.scatter(x=x1, y=x2)
```

```
x3 = np.random.randn(300)*50

_ = plt.scatter(x=x1, y=x2, s=x3)
```

```
_ = plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)
```

# Wyświetlanie obrazów

```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```
img = mpimg.imread('office.jpg')
img
```

```
img.shape
```

```
plt.imshow(img)
plt.axis('Off')
```

```
fig = plt.figure(figsize=(12, 10))
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.axis('Off')
plt.subplot(222)
plt.imshow(mpimg.imread('business.jpg'))
plt.axis('Off')
plt.subplot(223)
plt.imshow(mpimg.imread('children.jpg'))
plt.axis('Off')
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
plt.axis('Off')
```

```
fig = plt.figure(figsize=(12, 10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
  plt.subplot(140 + idx + 1)
  plt.imshow(mpimg.imread(img))
  plt.axis('Off')
```
